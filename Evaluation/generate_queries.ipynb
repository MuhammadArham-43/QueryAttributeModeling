```python
import pandas as pd
import os
import random
```

```python
DATA_DIR = "../Data/Toydata_w_new_att_final.xlsx"

df = pd.read_excel(DATA_DIR)

print(df.head())
```

```python
df.rename(columns={"Unnamed: 0": "id"}, inplace=True)
```

```python
unique_brands = df["manufacturer"].unique()
min_price = df["price_in_usd"].min()
max_price = df["price_in_usd"].max()


print(unique_brands, len(unique_brands))
print(min_price)
print(max_price)
```

```python
def get_random_brand():
    return random.choice(unique_brands)

def get_random_price_for_brand(brand):
    df_brand = df[df["manufacturer"] == brand]
    return random.choice(df_brand["price_in_usd"].tolist())

def get_random_age_for_brand(brand):
    df_brand = df[df["manufacturer"] == brand]
    return random.choice(df_brand["Preferred_Age"].tolist())

def get_random_rating_for_brand(brand):
    df_brand = df[df["manufacturer"] == brand]
    return random.choice(df_brand["star_rating"].tolist())

def get_random_product_for_brand(brand):
    df_brand = df[df["manufacturer"] == brand]
    return random.choice(df_brand["product_name"].tolist())
```

```python
def get_random_json():
    brand = get_random_brand()
    age = get_random_age_for_brand(brand)
    price = get_random_price_for_brand(brand)
    rating = get_random_rating_for_brand(brand)
    product = get_random_product_for_brand(brand)
    return {
        "brand": brand,
        "age": age,
        "price": price,
        "rating": rating,
        # "product": product
    }

def get_random_values():
    brand = get_random_brand() 
    price = get_random_price_for_brand(brand)
    age = get_random_age_for_brand(brand)
    rating = get_random_rating_for_brand(brand)
    return brand, price, age, rating
```

```python
brand = get_random_brand()
price = get_random_price_for_brand(brand)
age = get_random_age_for_brand(brand)
rating = get_random_rating_for_brand(brand)
product = get_random_product_for_brand(brand)
print(brand, price, age, rating, product)
```

```python
from openai import AzureOpenAI 

client = AzureOpenAI(
    api_key="",
    api_version="2024-02-15-preview",
    azure_endpoint="",
    azure_deployment="gpt-4o-large"
)
```

```python
LLM_PROMPT = """
Generate natural and human-like queries that someone might ask while searching for this product.
Make the query as human-like as possible who is searching for products in a website
Choose the attributes from the below list.
    brand:{brand}
    age:{age}
    price:{price}
    rating:{rating}
Follow the above given JSON to generate a query following the attributes specified in the JSON. Only output 5 queires as a numbered listand nothing else.
"""
```

```python
brand, age, price, rating = get_random_values()
print(brand, age, price, rating)
prompt = LLM_PROMPT.format(brand=brand, age=age, price=price, rating=rating)
print(prompt)
```

```python
response = client.chat.completions.create(
    model="gpt-4o-large",
    messages=[{"role": "user", "content": prompt}],
    temperature=0.5,
    max_tokens=1000
)

print(response.choices[0].message.content)
```

```python
import re
queries = re.findall(r'\d+\.\s*(.*)', response.choices[0].message.content)
print(queries)
```

```python
def _generate_queries():
    brand, age, price, rating = get_random_values()
    prompt = LLM_PROMPT.format(brand=brand, age=age, price=price, rating=rating)
    response = client.chat.completions.create(
        model="gpt-4o-large",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.5,
        max_tokens=1000
    )
    queries = (re.findall(r'\d+\.\s*(.*)', response.choices[0].message.content))
    return queries
```

```python
_generate_queries()
```

```python
from tqdm import tqdm
queries = []
```

```python
for _ in tqdm(range(100)):
    queries += _generate_queries()
queries = list(set(queries))
```

```python
queries = [query.replace("\"", "") for query in queries]
```
